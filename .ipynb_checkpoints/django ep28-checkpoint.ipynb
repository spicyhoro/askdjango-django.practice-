```python
image_url = 'http://www.v3wall.com/wallpaper/1920_1080/1004/1920_1080_20100420010355464809.jpg'
```

```python
import requests

image_binary = requests.get(image_url).content
```

```python
from PIL import Image
from pilkit.processors import Thumbnail

processor = Thumbnail(width=300, height=300)

image_cat = Image.open("cat.jpg")
thumb_img = processor.process(image_cat)
thumb_img = thumb_img.convert("RGBA")


thumb_img.save("cat300x300.png")

for quality in [100, 80, 60, 40, 20]:
   thumb_img.save("sample-300x300-{}.jpg".format(quality), quality=quality)
```
